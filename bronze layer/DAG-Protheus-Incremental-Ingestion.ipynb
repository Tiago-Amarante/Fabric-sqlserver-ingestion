```python
def split_list_into_n_parts(lst, n):
    """
    Splits a list into n nearly equal parts.
    :param lst: List to be split.
    :param n: Number of parts to split the list into.
    :return: A list of n lists, where each sublist contains the divided elements of the original list.
    """
    # Calculate the length of each part. The division operation "//" ensures an integer result
    part_len = len(lst) // n
    
    # Calculate the number of parts that need to be one element larger to distribute the remainder
    remainder = len(lst) % n
    
    # Initialize variables to keep track of the current index in the original list
    start = 0
    
    # Initialize the list that will hold all the parts
    parts = []
    
    # Generate each part
    for i in range(n):
        # Parts with an index < remainder will have one extra element
        end = start + part_len + (1 if i < remainder else 0)
        # Append the current part to the list of parts
        parts.append(lst[start:end])
        # Update the start index for the next part
        start = end
    
    return parts
```

```python
# Parse the JSON string
parsed_json = json.loads(protheus_tables)

# Extract the values of the "TABLE_NAME" keys
table_names = [item["TABLE_NAME"] for item in parsed_json]
```
