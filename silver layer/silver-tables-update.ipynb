```python
tables_names = tables_names.replace(']','').replace('[','').split(',')
```

```python
def df_to_string(df):
    column_values = [str(row['COLUMN_NAME']) for row in df.collect()]
    result_string = ', '.join(column_values)
    return result_string
```

```python
def get_table_schema(table):
    query = f'''
    SELECT COLUMN_NAME 
    FROM silver_protheus.tableschema_silver 
    WHERE TABLE_NAME = {table.upper()} 
    ORDER BY ORDINAL_POSITION
    '''
    df_schema = spark.sql(query)
    
    return df_to_string(df_schema).replace(', F1_ZJSONIN','')
```

```python
# generate the string to be queried
def create_query_str(table,columns_to_be_queried):
    table_name = table.replace('\'','').replace(' ','')
    query_string = f'''
    SELECT {columns_to_be_queried},
    INGESTION_DATE
    FROM silver_protheus.{table_name}
    '''
    return query_string
```

```python
# this cell is used to debug, keep it frozen
for table in tables_names:
    print(table)
    columns_to_be_queried = get_table_schema(table)
    # print('1')
    query_string = create_query_str(table,columns_to_be_queried)
    print(query_string)
    df = spark.sql(query_string)
    # print('3')
    # write_parquet_file(df,table)
```

```python
def write_parquet_file(df,table):


    table_name = f'{table}_silver'
    table_name = table_name.replace('\'','').replace(' ','')

    df.write \
    .option("overwriteSchema", "true") \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(table_name,mode='overwrite', ifNotExists=True)
```
